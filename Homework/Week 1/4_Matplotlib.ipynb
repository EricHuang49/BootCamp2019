```python
import numpy as np
from matplotlib import pyplot as plt
```

## Problem 1

```python
# Problem 1
def var_random_normal(n):
    x = np.random.normal(size=(n,n))
    y = np.mean(x,axis = 1)
    return np.var(y)

x = [var_random_normal(i) for i in np.arange(100,1100,100)]
plt.plot(np.arange(100,1100,100),x)
plt.xlabel('n')
plt.ylabel('var of mean of n random normal variables')
plt.show()
```

## Problem 2

```python
# Problem 2
def plot_fun(a,b):
    x = np.linspace(a,b,100)
    plt.plot(x,np.sin(x),label='sin')
    plt.plot(x,np.cos(x),label='cos')
    plt.plot(x,np.arctan(x),label='arctan')
    plt.legend()
    
plot_fun(-2*np.pi,2*np.pi)
```

## Problem 3

```python
# Problem 3
function_name = r'$f(x)=\frac{1}{x-1}$'

x = np.linspace(-2,0.9,5000)
plt.plot(x,1/(x-1),'m--',lw = 4)
x = np.linspace(1.1,6,5000)
plt.plot(x,1/(x-1),'m--',lw = 4,label=function_name)
plt.xlim(-2,6)
plt.ylim(-6,6)
plt.legend()
plt.show()
```

## Problem 4

```python
# Problem 4
fig,ax = plt.subplots(2,2)
x = np.linspace(0,2*np.pi,100)
ax[0,0].plot(x,np.sin(x),'g')
ax[0,0].set_title('sin(x)')

ax[0,1].plot(x,np.sin(2*x),'r--')
ax[0,1].set_title('sin(2x)')

ax[1,0].plot(x,2*np.sin(x),'b--')
ax[1,0].set_title('2sin(x)')

ax[1,1].plot(x,2*np.sin(2*x),'m.')
ax[1,1].set_title('2sin(2x)')

plt.suptitle('These are my bros!')
fig.subplots_adjust(top=0.88)
plt.tight_layout()
```
